```python
# ==============================
# YEAR-LONG LIQUIDITY PLANNER
# (robust date parsing)
# ==============================

import os
from dataclasses import dataclass
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA_DIR   = r"D:\liquidity_planning"
CSV_PATH   = os.path.join(DATA_DIR, "cashflows_year.csv")
HORIZON    = 365

@dataclass
class StressConfig:
    inflow_down: float = 0.30   # cut inflows by 30%
    outflow_up: float  = 0.20   # boost outflows by 20%
    hqla_haircut: float = 0.05  # 5% haircut on opening HQLA

def create_sample_yearly_csv(path: str) -> pd.DataFrame:
    rows = []
    # A
    for m in range(1, 13):
        rows.append(("InsureCoA", f"2025-{m:02d}-01", "inflow", 10000000, "premiums"))
        rows.append(("InsureCoA", f"2025-{m:02d}-05", "outflow", 2000000,  "operating_expenses"))
    for d in ["2025-03-15","2025-06-15","2025-09-15","2025-12-15"]:
        rows.append(("InsureCoA", d, "outflow", 7000000, "claims"))
    rows.append(("InsureCoA", "2025-07-01", "outflow", 10000000, "reinsurance_premium"))
    rows.append(("InsureCoA", "2025-06-30", "inflow",  5000000, "investment_income"))
    rows.append(("InsureCoA", "2025-12-31", "inflow",  5000000, "investment_income"))
    # B
    for m in range(1, 13):
        rows.append(("InsureCoB", f"2025-{m:02d}-01", "inflow", 8000000,  "premiums"))
        rows.append(("InsureCoB", f"2025-{m:02d}-10", "outflow", 1500000, "operating_expenses"))
    for d in ["2025-03-20","2025-06-20","2025-09-20","2025-12-20"]:
        rows.append(("InsureCoB", d, "outflow", 5000000, "claims"))
    rows.append(("InsureCoB", "2025-07-15", "outflow", 7000000, "reinsurance_premium"))

    df = pd.DataFrame(rows, columns=["company","date","kind","amount","label"])
    os.makedirs(os.path.dirname(path), exist_ok=True)
    df.to_csv(path, index=False)
    return df

def parse_dates_robust(series: pd.Series) -> pd.Series:
    """
    Handles both 'YYYY-MM-DD' and 'DD/MM/YYYY' (and mixes of them).
    Tries a few strategies: ISO  mixed  dayfirst.
    """
    s = series.astype(str).str.strip()
    # 1) Try fast ISO parse
    try:
        return pd.to_datetime(s, format="%Y-%m-%d", errors="raise")
    except Exception:
        pass
    # 2) Try mixed (Pandas 2.x), then with dayfirst
    try:
        return pd.to_datetime(s, format="mixed", errors="raise")
    except Exception:
        pass
    # 3) Try dayfirst (handles 30/06/2025 etc.)
    return pd.to_datetime(s, dayfirst=True, errors="raise")

def run_liquidity(cashflows: pd.DataFrame, hqla_start: float, stress: StressConfig, horizon_days: int):
    df = cashflows.copy()

    # Clean/validate
    for col in ["date","kind","amount"]:
        if col not in df.columns:
            raise ValueError(f"Missing column: {col}")

    df["date"] = parse_dates_robust(df["date"])
    df["kind"] = df["kind"].astype(str).str.strip().str.lower()
    bad = set(df["kind"]) - {"inflow","outflow"}
    if bad:
        raise ValueError(f"'kind' must be inflow/outflow; got {bad}")

    # numeric amounts (strip commas just in case)
    df["amount"] = df["amount"].astype(str).str.replace(",","",regex=False).astype(float)

    # Apply stress
    df.loc[df["kind"]=="inflow","stressed"]  = df["amount"]*(1-stress.inflow_down)
    df.loc[df["kind"]=="outflow","stressed"] = df["amount"]*(1+stress.outflow_up)

    # Daily aggregation
    daily = df.groupby(["date","kind"])["stressed"].sum().unstack(fill_value=0.0)
    daily["net"] = daily.get("inflow",0.0) - daily.get("outflow",0.0)

    # Build full horizon
    daily.index = pd.to_datetime(daily.index)
    start = daily.index.min()
    end   = start + pd.Timedelta(days=horizon_days-1)
    idx   = pd.date_range(start, end, freq="D")
    daily = daily.reindex(idx, fill_value=0.0)
    daily.index.name = "date"

    # Roll forward
    hqla_eff = hqla_start*(1-stress.hqla_haircut)
    opening  = hqla_eff
    rows = []
    for dt, row in daily.iterrows():
        net = float(row["net"])
        closing = opening + net
        rows.append((dt, row.get("inflow",0.0), row.get("outflow",0.0), net, opening, closing))
        opening = closing

    out = pd.DataFrame(rows, columns=["date","inflow","outflow","net","opening_liquidity","closing_liquidity"]).set_index("date")

    net_outflows_total = float(abs(np.minimum(out["net"],0.0).sum()))
    lcr = (hqla_eff/net_outflows_total) if net_outflows_total>0 else float("inf")

    return {
        "daily": out.reset_index(),
        "hqla_effective": float(hqla_eff),
        "net_outflows_total": net_outflows_total,
        "lcr": float(lcr),
        "min_closing_liquidity": float(out["closing_liquidity"].min()),
    }

# Ensure CSV exists
if (not os.path.exists(CSV_PATH)) or (os.path.getsize(CSV_PATH)==0):
    print("CSV missing/empty  creating sample...")
    create_sample_yearly_csv(CSV_PATH)

# Robust read (handles UTF-8 / BOM / UTF-16)
for enc in [None, "utf-8", "utf-8-sig", "utf-16"]:
    try:
        cashflows_all = pd.read_csv(CSV_PATH, encoding=enc)
        break
    except Exception:
        pass

stress = StressConfig()
hqla_start = 150_000_000

for company, df_co in cashflows_all.groupby("company"):
    print(f"\n===== {company} =====")
    res = run_liquidity(df_co, hqla_start=hqla_start, stress=stress, horizon_days=HORIZON)

    print(f"Effective HQLA (after haircut): {res['hqla_effective']:,.0f}")
    print(f"Stressed net outflows (1y):     {res['net_outflows_total']:,.0f}")
    print(f"LCR-like (HQLA/outflows):       {('' if res['lcr']==float('inf') else f'{res['lcr']:.2f}x')}")
    print(f"Min closing liquidity (1y):     {res['min_closing_liquidity']:,.0f}")

    # Plot
    daily = res["daily"]
    plt.figure(figsize=(11,4))
    plt.plot(daily["date"], daily["closing_liquidity"], marker="o")
    plt.title(f"Closing Liquidity  {company} (Stressed, 1 Year)")
    plt.xlabel("Date"); plt.ylabel("Closing Liquidity")
    plt.grid(True); plt.xticks(rotation=45); plt.tight_layout()
    plt.show()

    # Show full 365 rows in notebook
    pd.set_option("display.max_rows", None)
    display(daily)

    # Save to disk
    out_path = os.path.join(DATA_DIR, f"daily_liquidity_{company}_stressed.csv")
    daily.to_csv(out_path, index=False)
    print(f"Saved: {out_path}")
```

```python
# ==== PDF export: full working cell ====
import os
from reportlab.lib.pagesizes import A4
from reportlab.lib.units import cm
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image
from reportlab.lib.styles import getSampleStyleSheet
import matplotlib.pyplot as plt

def export_liquidity_report(company, results, output_dir):
    """
    Create a 1-page PDF with summary metrics + a chart for one company.
    Saves to output_dir/Liquidity_Report_<company>.pdf
    """
    pdf_path   = os.path.join(output_dir, f"Liquidity_Report_{company}.pdf")
    chart_path = os.path.join(output_dir, f"liquidity_chart_{company}.png")

    styles = getSampleStyleSheet()
    doc = SimpleDocTemplate(
        pdf_path, pagesize=A4,
        leftMargin=1.7*cm, rightMargin=1.7*cm,
        topMargin=1.7*cm, bottomMargin=1.7*cm
    )
    elements = []

    # Title & subtitle
    elements.append(Paragraph("Liquidity Stress Report", styles["Title"]))
    elements.append(Paragraph(f"Company: {company}", styles["Heading2"]))
    elements.append(Spacer(1, 0.5*cm))

    # Summary table
    summary = [
        ["Metric", "Value"],
        ["Effective HQLA (after haircut)", f"{results['hqla_effective']:,.0f}"],
        ["Stressed Net Outflows (1y)",      f"{results['net_outflows_total']:,.0f}"],
        ["LCR-like Ratio",                  "" if results["lcr"] == float("inf") else f"{results['lcr']:.2f}x"],
        ["Minimum Closing Liquidity",       f"{results['min_closing_liquidity']:,.0f}"],
    ]
    tbl = Table(summary, colWidths=[8*cm, 8*cm])
    tbl.setStyle(TableStyle([
        ("BACKGROUND", (0,0), (-1,0), colors.lightgrey),
        ("GRID", (0,0), (-1,-1), 0.5, colors.grey),
        ("ALIGN", (0,0), (-1,-1), "CENTER"),
        ("BOTTOMPADDING", (0,0), (-1,0), 6),
        ("TOPPADDING",    (0,0), (-1,0), 6),
    ]))
    elements.append(tbl)
    elements.append(Spacer(1, 1*cm))

    # Chart: save PNG then embed
    daily = results["daily"]
    plt.figure(figsize=(8,3))
    plt.plot(daily["date"], daily["closing_liquidity"], marker="o")
    plt.title(f"Closing Liquidity  {company} (Stressed, 1 Year)")
    plt.xlabel("Date"); plt.ylabel("Closing Liquidity")
    plt.grid(True); plt.tight_layout()
    plt.savefig(chart_path, bbox_inches="tight")
    plt.close()

    elements.append(Image(chart_path, width=15*cm, height=6*cm))

    # Build the PDF
    doc.build(elements)
    print(f"PDF saved: {pdf_path}")

# === Generate a PDF for each company ===
# Assumes your notebook already has: cashflows_all, run_liquidity, StressConfig defined
HORIZON = 365
stress = StressConfig()
hqla_start = 150_000_000
OUTPUT_DIR = DATA_DIR if 'DATA_DIR' in globals() else r"D:\liquidity_planning"

for company, df_co in cashflows_all.groupby("company"):
    res = run_liquidity(df_co, hqla_start=hqla_start, stress=stress, horizon_days=HORIZON)
    export_liquidity_report(company, res, OUTPUT_DIR)
```

```python
def export_liquidity_report(company, results, output_dir):
    import os
    from reportlab.lib.pagesizes import A4
    from reportlab.lib.units import cm
    from reportlab.lib import colors
    from reportlab.platypus import (
        SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image, PageBreak
    )
    from reportlab.lib.styles import getSampleStyleSheet
    import matplotlib.pyplot as plt

    pdf_path   = os.path.join(output_dir, f"Liquidity_Report_{company}.pdf")
    chart_path = os.path.join(output_dir, f"liquidity_chart_{company}.png")

    styles = getSampleStyleSheet()
    doc = SimpleDocTemplate(pdf_path, pagesize=A4,
                            leftMargin=1.7*cm, rightMargin=1.7*cm,
                            topMargin=1.7*cm, bottomMargin=1.7*cm)
    elements = []

    # Title
    elements.append(Paragraph("Liquidity Stress Report", styles["Title"]))
    elements.append(Paragraph(f"Company: {company}", styles["Heading2"]))
    elements.append(Spacer(1, 0.5*cm))

    # Summary table
    summary = [
        ["Metric", "Value"],
        ["Effective HQLA (after haircut)", f"{results['hqla_effective']:,.0f}"],
        ["Stressed Net Outflows (1y)", f"{results['net_outflows_total']:,.0f}"],
        ["LCR-like (HQLA / Outflows)", "" if results["lcr"]==float("inf") else f"{results['lcr']:.2f}x"],
        ["Minimum Closing Liquidity", f"{results['min_closing_liquidity']:,.0f}"],
    ]
    tbl = Table(summary, colWidths=[8*cm, 8*cm])
    tbl.setStyle(TableStyle([
        ("BACKGROUND", (0,0), (-1,0), colors.lightgrey),
        ("GRID", (0,0), (-1,-1), 0.25, colors.grey),
        ("ALIGN", (0,0), (-1,-1), "CENTER"),
    ]))
    elements.append(tbl)
    elements.append(Spacer(1, 1*cm))

    # Chart
    daily = results["daily"]
    plt.figure(figsize=(8,3))
    plt.plot(daily["date"], daily["closing_liquidity"], marker="o")
    plt.title(f"Closing Liquidity  {company} (Stressed, 1 Year)")
    plt.xlabel("Date"); plt.ylabel("Closing Liquidity")
    plt.grid(True); plt.tight_layout()
    plt.savefig(chart_path, bbox_inches="tight"); plt.close()
    elements.append(Image(chart_path, width=15*cm, height=6*cm))

    # Appendix: first + last 10 days of table
    elements.append(PageBreak())
    elements.append(Paragraph("Appendix: Daily Liquidity Table", styles["Heading2"]))
    head = daily.head(10)
    tail = daily.tail(10)

    def df_to_rows(df):
        rows = [list(df.columns)]
        rows += [list(map(lambda x: f"{x:,.0f}" if isinstance(x,(int,float)) else str(x), r)) for r in df.values]
        return rows

    rows = [[" FIRST 10 DAYS "]] + df_to_rows(head) + [[""], [" LAST 10 DAYS "]] + df_to_rows(tail)
    table = Table(rows, repeatRows=1, colWidths=[3*cm, 3*cm, 3*cm, 3*cm, 3*cm, 3*cm])
    table.setStyle(TableStyle([
        ("GRID", (0,0), (-1,-1), 0.25, colors.grey),
        ("BACKGROUND", (0,1), (-1,1), colors.whitesmoke),
    ]))
    elements.append(table)

    # Build PDF
    doc.build(elements)
    print(f"PDF saved: {pdf_path}")
```

```python
for company, df_co in cashflows_all.groupby("company"):
    res = run_liquidity(df_co, hqla_start=150_000_000, stress=StressConfig(), horizon_days=365)
    export_liquidity_report(company, res, DATA_DIR)
```
